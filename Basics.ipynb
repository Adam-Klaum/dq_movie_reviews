```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
movies = pd.read_csv('fandango_score_comparison.csv')
```

```python
movies
```

```python
ax = movies['Metacritic_norm_round'].hist(bins=5)
ax.set_xlim(0,5)
```

```python
ax = movies['Fandango_Stars'].hist(bins=5)
ax.set_xlim(0,5)
```

Looking at the histograms for Metacritic vs. Fandango it is obvious that Fandago's ratings are skewed quite high and are not at all a normal distribution.

```python
f_mean = movies['Fandango_Stars'].mean()
m_mean = movies['Metacritic_norm_round'].mean()
f_median = movies['Fandango_Stars'].median()
m_median = movies['Metacritic_norm_round'].median()
f_std = np.std(movies['Fandango_Stars'])
m_std = np.std(movies['Metacritic_norm_round'])
```

```python
stats = pd.DataFrame([['Fandango', f_mean, f_median, f_std],
                      ['Metacritic', m_mean, m_median, m_std]],
                     columns=['Provider', 'Mean', 'Median', 'Std. Dev'])
```

```python
stats
```

```python
plt.scatter(movies['Fandango_Stars'], movies['Metacritic_norm_round'])
```

```python
movies['fm_diff'] = movies['Metacritic_norm_round'] - movies['Fandango_Stars']
```

```python
movies['fm_diff'] = movies['fm_diff'].apply(lambda x: np.absolute(x))
```

```python
movies.sort_values(by=['fm_diff'], ascending=False, inplace=True)
```
